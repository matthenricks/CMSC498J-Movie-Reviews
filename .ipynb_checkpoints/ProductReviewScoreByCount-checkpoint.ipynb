```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
product_info_data = pd.DataFrame.from_csv("Data/Product_Info_Data.csv")
electronics = pd.DataFrame.from_csv("Data/RAW_Electronics_Review_Data.csv")
review_summaries = pd.DataFrame.from_csv("Data/Reviewer_Summaries.csv")
```

```python
print "review summaries:", review_summaries.keys()
print "product info:", product_info_data.keys()
print "electronics:", electronics.keys()
```

```python
plt.plot(product_info_data["Number_Of_Reviews"], product_info_data["Avg_Review_Score"])
```

### Standard Deviation Decreases

```python
p_group = product_info_data.groupby("Number_Of_Reviews")
p_group_avg = p_group.mean()
p_group_std = p_group.std()

plt.plot(p_group_avg.index, p_group_std["Avg_Review_Score"])
```

### Average Review Score Increases

```python
plt.plot(p_group_avg.index, p_group_avg["Avg_Review_Score"])
```
