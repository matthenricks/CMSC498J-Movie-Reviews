```python
import numpy as np
import pandas as pd
import math
```

```python
filtered_reviews = pd.DataFrame.from_csv("Data/Matt_Filtered_Electronics_Review_Data.csv")
print "review length: " + str(len(filtered_reviews))
```

```python
filtered_reviews.keys()
```

```python
user_groups = filtered_reviews[["Review_UserID", "Product_ID", "Review_Score"]].groupby("Review_UserID")
```

```python
self_dot = {}
for user in filtered_reviews.Review_UserID.unique():
    s1 = user_groups.get_group(user)[["Review_Score"]]
    s1 = s1 * s1
    self_dot[user] = sum(s1["Review_Score"])
```

```python
self_dot
```

```python
u1 = user_groups.get_group("A83G9QYT44X4F")[["Product_ID", "Review_Score"]]
u2 = user_groups.get_group("A1ZUJM8VAX5KFK")[["Product_ID", "Review_Score"]]
intersect = set(u1.Product_ID).intersection(set(u2.Product_ID))

dot_prod = 0
for p_id in intersect:
    s1 = float(u1[u1["Product_ID"] == p_id]["Review_Score"])
    s2 = float(u2[u2["Product_ID"] == p_id]["Review_Score"])
    dot_prod += s1 * s2

print dot_prod
```

```python
print "Cosine: ", str(dot_prod/math.sqrt(self_dot["A83G9QYT44X4F"]*self_dot["A1ZUJM8VAX5KFK"]))
```
